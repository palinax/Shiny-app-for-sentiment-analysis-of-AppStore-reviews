```r
install.packages("appler")
library(appler)

get_apple_reviews(
  id = 547702041,
  country = "us",
  all_results = FALSE,
  page_no = 2,
  sort_by = c("mostrecent", "mosthelpful")
)
```

```r

appstore_scraper <- function(app_id, n_reviews) {
  # Get app data
  app_data <- get_app_data(app_id)
  
  # Get app reviews
  app_reviews <- get_reviews(app_id, n_reviews)
  
  # Print app data and reviews
  print(app_data)
  print(app_reviews)
}
```

```r
app_id <- "547702041"
n_reviews <- 11
appstore_scraper(app_id, n_reviews)
```
